```python
import numpy as np
import matplotlib.pyplot as plt


def bar_plot_function(values, title, y_axis_name, change_scale=False):
  # Data for the bar plot
  labels = ['Softmax Baseline', 'JSD', 'JSD with Adaptive', 'Baseline', 'JSD', 'JSD with Adaptive']
  x_labels = ['Large', 'Large', 'Large', 'Fine-tune', 'Fine-tune', 'Fine-tune']
  colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#1f77b4', '#ff7f0e', '#2ca02c']

  # Number of bars in each group
  n_bars = 3

  # Creating the bar plot with model types as x-axis
  fig, ax = plt.subplots(figsize=(12, 8))

  bar_width = 0.2
  index = np.arange(2)  # We have 2 types of models: 'Large' and 'Fine-tune'

  # Plotting the bars for the different labels
  bars_none = ax.bar(index - bar_width, [values[0], values[3]], bar_width, color='#1f77b4', label='Softmax Baseline')
  bars_fixed = ax.bar(index, [values[1], values[4]], bar_width, color='#ff7f0e', label='JSD')
  bars_decaying = ax.bar(index + bar_width, [values[2], values[5]], bar_width, color='#2ca02c', label='JSD with Adaptive')

  # Adding labels and title
  ax.set_xlabel('Model Type', fontsize=14)
  ax.set_ylabel(y_axis_name, fontsize=14)
#   ax.set_title(title, fontsize=16, fontweight='bold')
  ax.set_xticks(index)
  ax.set_xticklabels(['T5-Large', 'T5-Large-Finetuned'], fontsize=12)
  ax.legend(title="Method", fontsize=12, title_fontsize=14)

  # Adding grid lines
  ax.grid(True, which='both', linestyle='--', linewidth=0.5, alpha=0.7)

  # Adding value labels on top of the bars
  def add_labels(bars):
    if change_scale:
      for bar in bars:
          height = bar.get_height()
          ax.annotate(f'{height:.2e}',
                      xy=(bar.get_x() + bar.get_width() / 2, height),
                      xytext=(0, 3),  # 3 points vertical offset
                      textcoords='offset points',
                      ha='center',
                      va='bottom',
                      fontsize=12)
    else:
      for bar in bars:
        height = bar.get_height()
        ax.annotate(f'{height:.2f}',
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords='offset points',
                    ha='center',
                    va='bottom',
                    fontsize=12)

  add_labels(bars_none)
  add_labels(bars_fixed)
  add_labels(bars_decaying)

  # Stretching the y-axis by a factor of 1.2
  max_height = max([bar.get_height() for bar in bars_none+bars_fixed+bars_decaying])
  ax.set_ylim(0, max_height * 1.2)

  # Adjust layout
  plt.tight_layout()

  plt.savefig(title.replace(" ", "_") + ".png")

  # Display the plot
  plt.show()

```

```python
# Data for the bar plot
## SQUAD peformance SNELLIUS
values = [92.446, 91.47, 91.34, 90.63,  90.59, 90.59]
# title = 'Squad: Performance of the models with regards to f1 with either no, fixed or decaying reduction'
# title = "\n".join(textwrap.wrap(title, width=80))
y_axis_name = "F1"
bar_plot_function(values, title, y_axis_name)
```
