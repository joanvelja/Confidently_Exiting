```python
import matplotlib.pyplot as plt

exit_min_layer_list = [18, 17, 15, 6]
average_exit_block_list_JDS = [19.242, 18.85, 18.16, 12.351]
average_exit_block_list_softmax = [20.188, 19.854, 19.439, 19.207]
average_exit_block_list_rew = [19.953, 19.599, 19.359, 8.62]
exit_normal = [24]

plt.plot(exit_min_layer_list, average_exit_block_list_JDS, marker='o', label='Average Exit Block with JDS', color='blue')
plt.plot(exit_min_layer_list, average_exit_block_list_softmax, marker='+', label='Average Exit Block with Softmax', color='red')
plt.plot(exit_min_layer_list, average_exit_block_list_rew, marker='x', label='Average Exit Block with Reweight cd', color='orange')
plt.grid(True)
plt.xlim(0, 25)
plt.xticks(exit_min_layer_list + exit_normal)  
plt.xlabel('Exit Min Layer')
plt.ylabel('Average Exit Block')
plt.title('Average Exit Block vs Exit Min Layer on SQuAD Dataset')
plt.legend()  # This will add a legend with the labels specified in the plot calls
plt.show()
```
